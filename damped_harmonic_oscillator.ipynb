## üßÆ The Beautiful Theory
The damped harmonic oscillator is like a poetry in physics, described by this elegant equation:

$$m\frac{d^2x}{dt^2} + c\frac{dx}{dt} + kx = 0$$

What's dancing in this equation? ü©∞
- m = mass (our main character)
- c = damping coefficient (the resistance)
- k = spring constant (the restoring force)
- x = position (where we are)
- dx/dt = velocity (how fast we're moving)
- d¬≤x/dt¬≤ = acceleration (how our speed changes)

### üîÑ State Space Magic
Here's where it gets interesting! We can transform our 2nd order ODE into two 1st order ODEs using the state space sorcery ‚ú®

Let's say $x$ be the position and $v = \dot{x}$ be the velocity.

And voil√†! Our system becomes this beautiful matrix:
$$\begin{bmatrix} \dot{x} \\ \dot{v} \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -\omega^2 & -\zeta \end{bmatrix} \begin{bmatrix} x \\ v \end{bmatrix}$$

The magical ingredients ü™Ñ:
- $\omega = \sqrt{\frac{k}{m}}$ (natural frequency - the system's heartbeat üíì)

- $\zeta = {\frac{c}{m}}$ (damping ratio - how quickly the dance fades üåô)

## üíª Let's Write Some Code!
Time to bring our mathematics to life with Python magic! üêç

```python
k = 1
c = 0.25
m = 1
omega_sq = k / m
zeta = c / m
```

```python
A = np.array([[0, 1], [-omega_sq, -zeta]])
A, A.shape
```

```python
x_0 = 1
v_0 = 0
init_state = np.array([[x_0], [v_0]])
init_state, init_state.shape
```

```python
dt = 0.001
T = 35
n_steps = int(T/dt)
t = np.linspace(0, T, n_steps)
current_state = init_state
states = np.zeros((2, n_steps))
for step in range(n_steps):
    states[:, step] = current_state[:, 0]
    derivative = A @ current_state
    current_state = current_state + derivative*dt
```

```python
# Plotting
fig = plt.figure(figsize=(12, 6))
fig.suptitle("Damped Harmonic Oscillator using Forward Euler Integration")
plt.subplot(121)
plt.plot(t, states[0, :], 'b-', label='Position')
plt.plot(t, states[1, :], 'r--', label='Velocity')
plt.grid(True)
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('State')
plt.title('Position and Velocity vs Time')

plt.subplot(122)
plt.plot(states[0, :], states[1, :], 'g-')
plt.grid(True)
plt.xlabel('Position')
plt.ylabel('Velocity')
plt.title('Phase Space')

plt.tight_layout()
plt.show()
```
